```python
import random
import google.generativeai as genai
from ipywidgets import widgets, Layout, VBox, HTML

# Configurações do modelo de IA generativa
from google.colab import userdata
api_key = userdata.get('SECRET_KEY')
genai.configure(api_key=api_key)

generation_config = {
    "candidate_count": 1,
    "temperature": 0.7,
}

safety_settings = {
    "HARASSMENT": "BLOCK_NONE",
    "HATE": "BLOCK_NONE",
    "SEXUAL": "BLOCK_NONE",
    "DANGEROUS": "BLOCK_NONE",
}

model = genai.GenerativeModel(model_name="gemini-1.0-pro",
                              generation_config=generation_config,
                              safety_settings=safety_settings)

chat = model.start_chat(history=[])

# Função para exibir mensagens do chatbot
def chatbot_message(msg):
    return widgets.HTML(value=f"<div style='text-align: left;'><p>{msg}</p></div>")

# Função para adicionar mensagens do chatbot à interface
def add_chatbot_message(msg):
    chatbot = chatbot_message(msg)
    chat_area.children += (chatbot,)

# Função para enviar mensagem do usuário e mostrar resposta do chatbot
def send_message(sender):
    user_msg = input_box.value.strip()
    if user_msg:
        add_message(user_msg)
        chatbot_response = chat.send_message(user_msg).text
        add_chatbot_message(chatbot_response)
        input_box.value = ""

# Função para adicionar mensagens do usuário à interface
def add_message(user_msg):
    user = widgets.HTML(value=f"<div style='text-align: left;'><p>{user_msg}</p></div>")
    chat_area.children += (user,)

# Interface do Chat
header = HTML("<h1 style='text-align:center;'>Chatbot IA Imersão Alura</h1><p style='text-align:center;'>Este chatbot é dedicado à conscientização sobre segurança cibernética. Sinta-se à vontade para fazer perguntas ou discutir qualquer coisa relacionada à segurança online!</p>")
chat_area = VBox()
input_box = widgets.Text(placeholder="Digite sua mensagem aqui...", layout=Layout(width='80%', margin='0px 0px 10px 0px'))
send_button = widgets.Button(description="Enviar", button_style="info")

input_box.on_submit(send_message)
send_button.on_click(send_message)

# Exemplos de frases para incentivar a conversa
frases_exemplo = [
    "O que é segurança da informação?",
    "Como posso proteger minha senha?",
    "Qual é a importância da autenticação de dois fatores?",
    "Como identificar um ataque de phishing?",
    "O que é engenharia social?",
    "Quais são os principais tipos de malware?",
    "Como posso proteger meu computador contra vírus?",
    "Quais são as melhores práticas de segurança cibernética?",
    "Como criar uma senha forte e segura?",
    "O que é criptografia e por que é importante?",
]

# Adicionar cabeçalho e caixa de entrada à área do chat
chat_area.children += (header,)

# Exibir os exemplos de frases
for frase in frases_exemplo:
    chat_area.children += (HTML(f"<p>{frase}</p>"),)

# Adicionar caixa de entrada e botão de envio à área do chat
chat_area.children += (input_box, send_button)

# Exibir a área do chat
display(chat_area)
```
